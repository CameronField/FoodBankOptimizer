```python
import pandas as pd
```

```python
county_risk_df = pd.read_csv("CountyRisk.csv", index_col=0)
city_df = pd.read_csv("MS_City_demand.csv", index_col=0)
```

```python
county_risk_df.head()
```

```python
city_df.head()
```

```python
# Drop the now unnecessary columns
city_df.drop(columns=['FI Population 2025', "CityWeight", "2025 Cost Per Meal"], inplace=True)
city_df.head()
```

```python
city_weights = pd.read_csv("City_Weight.csv")
city_weights.head()
```

```python
# Merge city_df with city_weights on 'City'
city_df = city_df.merge(city_weights, on='City')

# Reset the index of county_risk_df to turn the 'County' index into a column
county_risk_df.reset_index(inplace=True)

# Now perform the merge operation
city_df = city_df.merge(county_risk_df[['County', '2025 FI Population', '2025 Cost Per Meal']], on='County')

# Calculate the FI Population 2025 for each city
city_df['FI Population 2025'] = city_df['2025 FI Population'] * city_df['CityWeight']

# Drop the now unnecessary columns
city_df.drop(columns=['2025 FI Population'], inplace=True)

# Display the updated city_df (optional)
print(city_df.head())

# Save the updated dataframe to a new file (optional)
city_df.to_csv('MS_City_Demand.csv', index=False)
```
